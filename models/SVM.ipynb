```python
X = np.concatenate((x_train, x_val), axis=0)
y = np.concatenate((y_train, y_val), axis=0)
train_sizes, train_scores, val_scores = learning_curve(clf, X, y, cv=5)

train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)
val_scores_std = np.std(val_scores, axis=1)

plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std, alpha=0.1,
                 color="r")
plt.fill_between(train_sizes, val_scores_mean - val_scores_std,
                 val_scores_mean + val_scores_std, alpha=0.1,
                 color="g")
plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
         label="Training score")
plt.plot(train_sizes, val_scores_mean, 'o-', color="g",
         label="Cross-validation score")
plt.xlabel('Number of Training Samples')
plt.ylabel('Accuracy Score')
plt.legend(loc="best")
plt.show()
```

### SVM Graphs

```python
values= np.arange(0.01,0.1,0.01)
train_errors = []
valid_errors = []

for value in values:
    clf = svm.SVC(kernel="poly",gamma=value, C=0.01)
    clf.fit(x_train, y_train)
    predictions=clf.predict(x_val)
    train_error = 1.0 - clf.score(x_train,y_train)
    valid_error = 1.0 - clf.score(x_val, y_val)
    train_errors.append(train_error)
    valid_errors.append(valid_error)

# Plot the train and validation errors vs. gamma
plt.plot(values, train_errors, label='Train error')
plt.plot(values, valid_errors, label='Validation error')
plt.xlabel('Gamma with Poly Kernel')
plt.ylabel('Classification error')
plt.title('Train error vs Validation error')
plt.legend()
plt.show()
```
