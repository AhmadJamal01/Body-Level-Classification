```python
cs = []
train_err = []
val_err = []

for i in range(0,2000,1):
    # Training.
    cs_i = i
    if i == 0:
        cs_i = 1

    rforest_model = RandomForestClassifier(n_estimators = cs_i, criterion="entropy", max_features=None)
    rforest_model.fit(x_train, y_train)

    # Prediction
    train_pred = pd.DataFrame(rforest_model.predict(x_train))
    val_pred = pd.DataFrame(rforest_model.predict(x_val))

    cs.append(cs_i)
    train_err.append(1.0-accuracy_score(y_train,train_pred))
    val_err.append(1.0-accuracy_score(y_val,val_pred))
```
